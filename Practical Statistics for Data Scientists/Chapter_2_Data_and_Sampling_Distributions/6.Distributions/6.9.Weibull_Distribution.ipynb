# Weibull Distribution

## Overview

In many cases, the event rate does not remain constant over time. If the period over which it changes is much longer than the typical interval between events, there is no problem; you just subdivide the analysis into time segments where rates are relatively constant, as mentioned before. If, however, the event rate changes over the time of the interval, the exponential (or Poisson) distributions are no longer useful. This is likely to be the case in mechanical failure—the risk of failure increases as time goes by.  The Weibull distribution is an extension of the exponential distribution in which the event rate is allowed to change, as specified by a shape parameter, β. If β > 1, the probability of an event increases over time; if β < 1, the probability decreases. Because the Weibull distribution is used with time-to-failure analysis instead of event rate, the second parameter is expressed in terms of characteristic life, rather than in terms of the rate of events per interval. The symbol used is η, the Greek letter eta. It is also called the scale parameter. With the Weibull, the estimation task now includes estimation of both parameters, β and η. The Weibull distribution is widely used in reliability analysis and survival analysis, particularly for modeling time-to-failure data. This distribution allows for variable event rates, making it superior for applications where the event rate changes over time.

> **Key Concepts**
>
> - **Shape Parameter (β)**: Indicates how the rate of failure changes over time.
> - **Scale Parameter (η)**: Also known as the characteristic life, it represents the scale of the time-to-failure distribution.

## Estimating Failure Rates

Estimating the failure rate for rare events like aircraft engine failures poses unique challenges due to sparse data. The Weibull distribution's flexibility allows for modeling these scenarios even with limited or no prior data by adjusting the shape and scale parameters to fit observed lifetimes.

### Practical Example

If no engine failures have occurred over a significant period, we can infer that the failure rate is not as frequent as once per hour. This kind of reasoning helps in setting realistic expectations and preparing for potential risks more accurately.

## Mathematical Representation

The probability density function of the Weibull distribution is:
\[ f(x; \beta, \eta) = \frac{\beta}{\eta} \left( \frac{x}{\eta} \right)^{\beta - 1} e^{-\left( \frac{x}{\eta} \right)^\beta} \]
where \( x \) is the time until an event occurs, \( \beta \) affects the distribution's shape, and \( \eta \) is the scale parameter reflecting the typical lifespan.

## Visualization and Practical Application

Let's visualize the Weibull distribution to see how it can predict the lifespan of machinery or the time until the next failure.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

sns.set(style="whitegrid")

weibull_data = stats.weibull_min.rvs(1.5, scale=5000, size=1000)

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

sns.histplot(weibull_data, bins=30, kde=True, color='blue', ax=axs[0])
axs[0].set_title('Weibull Distribution')
axs[0].set_xlabel('Time to Failure')
axs[0].set_ylabel('Frequency')

stats.probplot(weibull_data, dist="norm", plot=axs[1])
axs[1].get_lines()[1].set_color('grey')
axs[1].set_title('QQ-Plot for Weibull Distribution')

plt.tight_layout()
plt.show() 
```

## Plot Explanations
Histogram: Illustrates the distribution of lifetimes, showing how the probability of failure changes over time.
QQ-Plot: Highlights the non-normal characteristics of the distribution, especially its tail behavior.

## Relationship with Other Distributions
The Weibull distribution builds on the Exponential distribution by incorporating a shape parameter, allowing for a more nuanced modeling of event timing, particularly useful in assessing reliability and predicting failure times.
