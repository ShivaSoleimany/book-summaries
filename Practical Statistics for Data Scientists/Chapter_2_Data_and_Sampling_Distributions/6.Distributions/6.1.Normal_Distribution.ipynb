# Normal Distribution

The normal distribution, recognized by its characteristic bell shape, plays a pivotal role in traditional statistics. Its prevalence in the distribution of sample statistics has facilitated the development of mathematical models and formulas used to approximate statistical distributions.

### Statistical Significance of the Normal Distribution

Within the normal distribution, approximately 68% of data falls within one standard deviation from the mean, and 95% falls within two standard deviations. These properties make the normal distribution an invaluable tool for statistical analysis and hypothesis testing.

### Common Misconceptions and Practical Applications

It is a widespread misconception that the term "normal distribution" implies most real-world data naturally follows this distribution. Contrary to this belief, raw data in many scientific and business applications do not exhibit normal distribution characteristics. However, the normal distribution is important because many statistical methods rely on the assumption that as sample sizes increase, the distribution of sample means or other statistics tends toward normality. This principle underpins many inferential statistics techniques, making assumptions of normality particularly useful when empirical distributions or bootstrap methods are not viable options.

The value of the normal distribution in statistical analysis primarily stems from its theoretical properties, which simplify calculations and predictions in many contexts where direct measurements are not feasible.

>## Essential Concepts of Normal Distribution
>- **Error:** The discrepancy between an observed data point and its expected (predicted or average) value.
>- **Standardize:** The process of adjusting data by subtracting the mean and dividing by the standard deviation.
>- **z-score:** A metric obtained by standardizing a data point, indicating how many standard deviations it is from the mean.
>- **Standard normal:** A specific type of normal distribution where the mean is 0 and the standard deviation is 1.
>- **QQ-Plot:** A graphical tool used to assess if a dataset follows a particular theoretical distribution, such as the normal distribution.


# Standard Normal and QQ-Plots

## Understanding Standard Normal Distribution

A **standard normal distribution** is a key concept in statistics, characterized by a mean of zero and a standard deviation of one. Data points in this distribution are expressed in units known as z-scores, which represent how many standard deviations a point is from the mean. To transform data into this scale—often referred to as normalization or standardization—you subtract the mean of the dataset and then divide by the standard deviation.

## QQ-Plots

**QQ-Plots (Quantile-Quantile Plots)** are graphical tools used to assess if a dataset follows a given theoretical distribution, such as the normal distribution. In practice, QQ-Plots:
- Calculate the z-scores (standard scores) of the sample data.
- Plot these z-scores on the y-axis against the expected quantiles of the normal distribution on the x-axis.
- Check for alignment with the diagonal line; if the sample data points largely adhere to this line, the sample can be considered approximately normally distributed.

### Practical Implementation with Python

Below is a Python script that demonstrates how to generate a standard normal sample and visualize it with a QQ-Plot to assess its normality. The code uses stats.norm.rvs to generate 100 random values from a standard normal distribution. The stats.probplot function from SciPy is utilized to generate the QQ-Plot, mapping the quantiles of the sample against the quantiles of a theoretical normal distribution. The resulting plot displays the sample quantiles against the theoretical quantiles. A grey line shows the ideal position where the sample quantiles would lie if they perfectly followed a normal distribution.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

norm_sample = stats.norm.rvs(size=100)

fig, ax = plt.subplots(figsize=(4, 4))
stats.probplot(norm_sample, plot=ax)
ax.get_lines()[1].set_color('grey') 
plt.show()
```
