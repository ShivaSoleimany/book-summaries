# Exponential Distribution

## Overview

The Exponential distribution is primarily used to model the time between events in a Poisson process, making it ideal for scenarios like the time between visits to a website or the intervals between cars passing a checkpoint.

> **Key Concepts**
>
> - **Lambda (Î»)**: The rate at which events occur, identical to the rate in the Poisson distribution.
> - **Exponential Distribution**: Describes the times between consecutive events in a Poisson process.

## Mathematical Representation

The formula for the probability density function of the Exponential distribution is given by:
\[ f(x; \lambda) = \lambda e^{-\lambda x} \]
where \( x \) is the time between events, and \( \lambda \) is the rate parameter, which is the reciprocal of the mean duration between events.

## Visualization and Practical Application

We will visualize the Exponential distribution to understand the behavior of intervals between successive events and how this can be applied, for example, in predicting the waiting times in customer service.

### Python Code for Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

sns.set(style="whitegrid")

exp_data = stats.expon.rvs(scale=1/0.2, size=1000)

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

sns.histplot(exp_data, bins=30, kde=True, color='blue', ax=axs[0])
axs[0].set_title('Exponential Distribution')
axs[0].set_xlabel('Time between Events')
axs[0].set_ylabel('Frequency')

stats.probplot(exp_data, dist="norm", plot=axs[1])
axs[1].get_lines()[1].set_color('grey')
axs[1].set_title('QQ-Plot for Exponential Distribution')

plt.tight_layout()
plt.show()
```

## Plot Explanations
Histogra## m: Displays the time intervals between events, typical of memoryless property.

QQ-Plot: Shows how the distribution's tail behavior differs significantly from that of a normal distribution.

## Relationship with Other Distributions
The Exponential distribution is often considered in conjunction with the Poisson distribution to provide a complete picture of the behavior of random events over time.
