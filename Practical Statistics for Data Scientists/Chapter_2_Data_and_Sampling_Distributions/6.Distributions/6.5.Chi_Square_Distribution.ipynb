#### Chi-Square Distribution

An important idea in statistics is departure from expectation, especially with respect to category counts. Expectation is defined loosely as nothing unusual or of note in the data (e.g., no correlation between variables or predictable patterns). This is also termed the null hypothesis or null model (see “The Null Hypothesis” on page 94). For example, you might want to test whether one variable (say, a row variable representing gender) is independent of another (say, a column variable representing “was promoted in job”), and you have counts of the number in each of the cells of the data table. The statistic that measures the extent to which results depart from the null expectation of independence is the chi-square statistic.


 It is the difference between the observed and expected values, divided by the square root of the expected value, then summed across all categories. This process standardizes the statistic so it can be compared to a reference distribution. A more general way of putting this is to note that the chi-square statistic is a measure of the extent to which a set of observed values “fits” a specified distribution (a “goodness-of-fit” test). It is useful for determining whether multiple treatments (an “A/B/C... test”) differ from one another in their effects.

The chi-square distribution is the distribution of the chi-square statistic under repeated resampled draws from the null model—see “Chi-Square Test” on page 124. The calculation of a value for a set of counts indicates that they closely follow the expected distribution. A high chi-square indicates that they differ markedly from what is expected. There are a variety of chi-square distributions associated with different degrees of freedom (e.g., number of observations—see “Degrees of Freedom” on page 116).

##### Key Ideas

- The chi-square distribution is typically concerned with counts of subjects or items falling into categories.
- The chi-square statistic measures the extent of departure from what you would expect in a null model.


# In-depth Exploration of the Chi-Square Distribution

The Chi-Square distribution is pivotal in assessing how observed results compare to expected outcomes under specific hypotheses, typically involving categorical data. This distribution is frequently employed in tests of independence and goodness-of-fit.

## Understanding Chi-Square Distribution

### Concept and Application

The Chi-Square statistic is crucial for evaluating if observed categorical data deviate from expected outcomes, assuming no relationship exists between variables in a null model. For example, in analyzing whether there is a relationship between two categorical variables such as gender (row variable) and promotion at work (column variable), the Chi-Square statistic can quantify the independence between these variables.

> **Key Concepts**
>
> - **Chi-Square Statistic**: Measures the degree to which observed frequencies differ from expected frequencies under the null hypothesis.
> - **Null Hypothesis**: A baseline assumption that there is no association between the variables being tested.
> - **Degrees of Freedom**: Typically, the number of categories minus one, adjusted for constraints such as the totals in each row and column.

### Chi-Square Statistic Formula


The computation involves each category's observed (\(O\)) and expected (\(E\)) frequencies:
$$\chi^2 = \sum \frac{(O - E)^2}{E}$$
This formula calculates the sum of the squared difference between observed and expected frequencies, divided by the expected frequency, across all categories.

### Visualization with Python

We'll now visualize the Chi-Square distribution to better understand its behavior and compare it to a normal distribution using QQ plots. These visualizations will demonstrate how the distribution behaves with different degrees of freedom.

#### Preparation for Plotting

Below, we'll see a histogram of the Chi-Square distribution along with a QQ plot that compares this distribution to a theoretical normal distribution. These plots will highlight the fit and deviations of observed data points.


````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

sns.set(style="whitegrid")

df = 10
chi_square_data = np.random.chisquare(df, size=1000)

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

sns.histplot(chi_square_data, bins=30, kde=True, color='blue', ax=axs[0])
axs[0].set_title('Chi-Square Distribution')
axs[0].set_xlabel('Chi-Square Values')
axs[0].set_ylabel('Frequency')

stats.probplot(chi_square_data, dist="norm", plot=axs[1])
axs[1].get_lines()[1].set_color('grey') 
axs[1].set_title('QQ-Plot for Chi-Square Distribution')

plt.tight_layout()
plt.show()
````
