#### Poisson and Related Distributions

Many processes produce events randomly at a given overall rate—visitors arriving at a website, or cars arriving at a toll plaza (events spread over time); imperfections in a square meter of fabric, or typos per 100 lines of code (events spread over space).

##### Key Terms for Poisson and Related Distributions

- **Lambda**: The rate (per unit of time or space) at which events occur.
- **Poisson distribution**: The frequency distribution of the number of events in sampled units of time or space.
- **Exponential distribution**: The frequency distribution of the time or distance from one event to the next event.
- **Weibull distribution**: A generalized version of the exponential distribution in which the event rate is allowed to shift over time.

From prior aggregate data (for example, number of flu infections per year), we can estimate the average number of events per unit of time or space (e.g., infections per day, or per census unit). We might also want to know how different this might be when addressing queuing questions such as “How much capacity do we need to be 95% sure of fully processing the internet traffic that arrives on a server in any five-second period?”

The key parameter in a Poisson distribution is λ, or lambda. This is the mean number of events that occurs in a specified interval of time or space. The variance for a Poisson distribution is also λ.

A common technique is to generate random numbers from a Poisson distribution as part of a queuing simulation. The rpois function in R does this, taking only two arguments—the quantity of random numbers sought, and lambda:

# Poisson Distribution

## Overview

The Poisson distribution models the number of discrete events within a given time or space, making it essential for understanding phenomena such as website visitor counts or typos in lines of code.

> **Key Concepts**
>
> - **Lambda (λ)**: The average rate of events per unit time or space.
> - **Poisson Distribution**: Models the frequency of a given number of events happening within a fixed interval.

## Mathematical Representation

The probability of observing exactly \( k \) events in an interval is given by:
\[ P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!} \]

## Visualizations

Below, we will visualize the Poisson distribution and compare its empirical quantiles to those of a theoretical normal distribution to understand its behavior.

### Python Code for Visualization

This code will generate 1000 random numbers from a Poisson distribution with λ = 2. For example, if incoming customer service calls average two per minute, this code will simulate 1000 minutes, returning the number of calls in each of those 1000 minutes.

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

sns.set(style="whitegrid")

poisson_data = stats.poisson.rvs(mu=2, size=1000)

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

sns.histplot(poisson_data, bins=30, kde=True, color='blue', ax=axs[0])
axs[0].set_title('Poisson Distribution')
axs[0].set_xlabel('Number of Events')
axs[0].set_ylabel('Frequency')

stats.probplot(poisson_data, dist="norm", plot=axs[1])
axs[1].get_lines()[1].set_color('grey')
axs[1].set_title('QQ-Plot for Poisson Distribution')

plt.tight_layout()
plt.show()
````

## Plot Explanations
Histogram: Shows the frequency of events, illustrating the typical skewness of the Poisson distribution.

QQ-Plot: Compares the empirical quantiles of the distribution to those of a normal distribution, highlighting deviations.

## Relationship with Other Distributions

The Poisson distribution often serves as a precursor to studying the Exponential distribution, which describes the time between Poisson events.
